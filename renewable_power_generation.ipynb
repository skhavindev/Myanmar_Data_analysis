## Chindwin river [station:hktami]

```python
import pandas as pd

file_path = 'GRDC_myanmar[Water_Discharge]/GRDC_myanmar/2260100_Q_Day.Cmd.txt'
data = pd.read_csv(file_path, delimiter=';', skiprows=36, encoding='ISO-8859-1')

data['hh:mm'] = data['hh:mm'].replace('--:--', '00:00')

data['DateTime'] = pd.to_datetime(data['YYYY-MM-DD'] + ' ' + data['hh:mm'], dayfirst=True, errors='coerce')


data = data.drop(columns=['YYYY-MM-DD', 'hh:mm'])


data.columns = ['Value', 'DateTime']


import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot
sns.lineplot(x='DateTime', y='Value', data=data)

# Add labels and a title
plt.xlabel('Date')
plt.ylabel('Water Discharge (mÂ³/s)')
plt.title('Water Discharge Over Time')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a histogram to visualize the distribution of water discharge values
sns.histplot(data['Value'], bins=30, kde=True)

# Add labels and a title
plt.xlabel('Water Discharge (mÂ³/s)')
plt.ylabel('Frequency')
plt.title('Distribution of Water Discharge Values')

# Show the plot
plt.tight_layout()
plt.show()

# Resample data by month, calculating the mean water discharge for each month
monthly_data = data.resample('ME', on='DateTime').mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot for the resampled monthly data
sns.lineplot(x=monthly_data.index, y='Value', data=monthly_data)

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Mean Water Discharge (mÂ³/s)')
plt.title('Average Monthly Water Discharge')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Resample data by year, calculating the mean water discharge for each year
yearly_data = data.resample('YE', on='DateTime').mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot for the resampled yearly data
sns.lineplot(x=yearly_data.index, y='Value', data=yearly_data)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Mean Water Discharge (mÂ³/s)')
plt.title('Average Yearly Water Discharge')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Create a new column for the year
data['Year'] = data['DateTime'].dt.year

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a boxplot to visualize the yearly variations in water discharge
sns.boxplot(x='Year', y='Value', data=data)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Water Discharge (mÂ³/s)')
plt.title('Yearly Water Discharge Variations')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()


```

```python
import pandas as pd

file_path = 'GRDC_myanmar[Water_Discharge]/GRDC_myanmar/2260600_Q_Day.Cmd.txt'
data = pd.read_csv(file_path, delimiter=';', skiprows=36, encoding='ISO-8859-1')

data['hh:mm'] = data['hh:mm'].replace('--:--', '00:00')

data['DateTime'] = pd.to_datetime(data['YYYY-MM-DD'] + ' ' + data['hh:mm'], dayfirst=True, errors='coerce')

data = data.drop(columns=['YYYY-MM-DD', 'hh:mm'])

data.columns = ['Value', 'DateTime']

import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot
sns.lineplot(x='DateTime', y='Value', data=data)

# Add labels and a title
plt.xlabel('Date')
plt.ylabel('Water Discharge (mÂ³/s)')
plt.title('Water Discharge Over Time')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a histogram to visualize the distribution of water discharge values
sns.histplot(data['Value'], bins=30, kde=True)

# Add labels and a title
plt.xlabel('Water Discharge (mÂ³/s)')
plt.ylabel('Frequency')
plt.title('Distribution of Water Discharge Values')

# Show the plot
plt.tight_layout()
plt.show()

# Resample data by month, calculating the mean water discharge for each month
monthly_data = data.resample('ME', on='DateTime').mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot for the resampled monthly data
sns.lineplot(x=monthly_data.index, y='Value', data=monthly_data)

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Mean Water Discharge (mÂ³/s)')
plt.title('Average Monthly Water Discharge')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Resample data by year, calculating the mean water discharge for each year
yearly_data = data.resample('YE', on='DateTime').mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a time series plot for the resampled yearly data
sns.lineplot(x=yearly_data.index, y='Value', data=yearly_data)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Mean Water Discharge (mÂ³/s)')
plt.title('Average Yearly Water Discharge')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Create a new column for the year
data['Year'] = data['DateTime'].dt.year

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a boxplot to visualize the yearly variations in water discharge
sns.boxplot(x='Year', y='Value', data=data)

# Add labels and a title
plt.xlabel('Year')
plt.ylabel('Water Discharge (mÂ³/s)')
plt.title('Yearly Water Discharge Variations')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```
